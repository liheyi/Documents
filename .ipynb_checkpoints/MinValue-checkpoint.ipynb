```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.mpl_style', 'default')
plt.rcParams['figure.figsize'] = (15, 5)

a = np.linspace(1,6,100000)
b = [6/x for x in a]
s = map(lambda i,j:(18 + i ** 2) * (8 + j ** 2)/2, a, b)

df = pd.DataFrame({'a': a, 'b': b, 'size': s})
df
```

```python
df['size'].plot()
```
