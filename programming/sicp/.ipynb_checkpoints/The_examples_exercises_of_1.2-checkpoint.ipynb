```python
# use generators to iteratively return 
# the sum of the first n Fibonacci numbers, starting from 0 

def fib(n):
    def fib_iter(n):
        a, b = 1, 0
        for _ in xrange(1,n):
            yield b
            a, b = a + b, a
    return sum(v for v in fib_iter(n))

fib(20)
#format(fib(20), ',d')
```

```python
value_of_coins={1:1, 2:5, 3:10, 4:25, 5:50}

def cc(amount,kinds_of_coins):
    if 0 == amount:
        return 1
    elif amount < 0 or 0 == kinds_of_coins:
        return 0
    else:
        return cc(amount, kinds_of_coins - 1) + \
                cc(amount - value_of_coins[kinds_of_coins], kinds_of_coins)

def count_change(amount):
    return cc(amount, 5)

count_change(100)
```

### II.exponent2

We can compute exponentials in fewer steps by using successive squaring.  
For instance, rather than computing $b^8$ as  
$$b*(b*(b*(b*(b*(b*(b*b))))))$$

we can compute it using three multiplications:  
$$b^2 = b * b,$$
$$b^4 = b^2 * b^2,$$
$$b^8 = b^4 * b^4.$$

This method works fine for exponents that are powers of 2. We can also take advantage of successive squaring in computing exponentials in general if we use the rule:

$b^n = (b^(n/2))^2$   if n is even,  
$b^n = b * b^(n-1)$    if n is odd.

#### A.Described by shceme from 1.2/exponent2_by_recursion.scm

```python
def fast_expt(b, n):
    if 0 == n:
        return 1
    elif 0 == n % 2:
        return int(pow(fast_expt(b, n / 2), 2))
    else:
        return b * fast_expt(b, n - 1)
    
fast_expt(3, 10)
```

## 1.2.6 Example: Testing for Primality

### A.Searching for divisors

### Answer:

根据素数的定义,如果一个正整数n为素数,那么n的因数只有1和n本身.  
因此我们通过寻找正整数n的大于1的最小因数,如果寻找到的最小因数就是n,则能确定n为素数.

为了找到正整数n的大于1的最小因数,我们需要定义一个找正整数n的因数的过程find_divisor.

其实,为了找到正整数n的大于1的最小因数,我们只需要通过过程find_divisor从2开始寻找第一个因数即可.  
1.我们可以测试从2开始一直到n的所有整数,如果2到n-1之间的某个整数能够整除n,那么n的最小因数小于n,则n不是素数;  
2.同时有这样的事实,如果n不是素数,则一定存在因数小于或者等于sqrt(n);意味着我们只需要测试1到sqrt(n)之间的整数即可.

将上面的代码片段简单的组合在一起就可以验证素数,我们将其写成块结构,如下:

```python
cat 1.2/prime_by_search_divisor1.scm
```

#### Python2.7

```python
def prime(n):
    
    def find_divisor(n, test_divisor):
        if pow(test_divisor, 2) > n:
            return n
        elif 0 == n % test_divisor:
            return test_divisor
        else:
            return find_divisor(n, test_divisor + 1)
        
    def smallest_divisor(n):
        return find_divisor(n, 2)
    
    if n == smallest_divisor(n):
        return True
    else:
        return False
```

### B.The Fermat test

### Answer:

To implement the Fermat test, we need a procedure that computes the exponential of a number modulo another number:

The Fermat test is performed by choosing at random a number a between 1 and n-􀀀1 inclusive and checking whether the remainder modulo n of the $a^n$ is equal to a.

The following procedure runs the test a given number of times, as specified by a parameter. Its value is true if the test succeeds every time, and false otherwise.

#### Scheme

```python
cat 1.2/prime_by_fermat_test.scm
```

#### Python2.7

```python
def fast_prime(n, times):
    
    def expmod(base, expo, m):
        if 0 == expo:
            return 1
        elif 0 == expo % 2:
            return pow(expmod(base, expo / 2, m), 2) % m
        else:
            return (base * expmod(base, expo - 1, m)) % m
        
    def fermat_test(n):
        from random import randint
        def try_it(a):
            return a == expmod(a, n, n)
        return try_it(randint(1, n - 1))
    
    if 0 == times:
        return True
    elif fermat_test(n):
        return fast_prime(n, times - 1)
    else:
        return False
```

```python
fast_prime(131, 500)
```

```python
fast_prime(9999991, 100)
```

```python
# Althouth times = 1
fast_prime(pow(2, 127) - 1, 100)
```
